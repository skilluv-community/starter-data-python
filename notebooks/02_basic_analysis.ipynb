# 02 — Basic analysis

Descriptive statistics and simple visualisations with matplotlib and seaborn.

*Français : stats descriptives et visualisations simples.*

```python
import sys
from pathlib import Path

sys.path.insert(0, str(Path.cwd().parent))

import matplotlib.pyplot as plt
import seaborn as sns

from src.utils import load_csv, summary_stats, top_n

df = load_csv("../data/raw/population_africa_sample.csv")
df.head()
```

```python
by_region = summary_stats(df, "region", "population_millions")
by_region
```

```python
top = top_n(df, "population_millions", n=5)

sns.set_theme(style="darkgrid")
fig, ax = plt.subplots(figsize=(8, 4))
sns.barplot(data=top, x="country", y="population_millions", ax=ax)
ax.set_title("Top 5 by population — sample data")
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
sns.scatterplot(data=df, x="population_millions", y="gdp_per_capita_usd", hue="region", ax=ax)
ax.set_title("Population vs GDP per capita")
plt.tight_layout()
plt.show()
```

# 04 — ML intro

A minimal scikit-learn pipeline: split, fit a linear regression predicting GDP per capita from population.

*Français : pipeline sklearn minimale — régression linéaire GDP ~ population.*

```python
import sys
from pathlib import Path
sys.path.insert(0, str(Path.cwd().parent))

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
from src.utils import load_csv

df = load_csv('../data/raw/population_africa_sample.csv')
X = df[['population_millions']].to_numpy()
y = df['gdp_per_capita_usd'].to_numpy()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression().fit(X_train, y_train)
y_pred = model.predict(X_test)
print('MAE:', mean_absolute_error(y_test, y_pred))
print('R2:', r2_score(y_test, y_pred))
print('coef:', model.coef_, 'intercept:', model.intercept_)
```

> With a 17-row toy dataset this model is essentially meaningless — the point is to show the sklearn API, not to draw conclusions.
>
> *Avec 17 lignes ce modèle n'a aucune valeur — l'objectif est de montrer l'API sklearn, pas de tirer des conclusions.*

# 03 — DuckDB queries

Query the same CSV directly with SQL via DuckDB — no import needed.

*Français : requêter le CSV directement en SQL via DuckDB.*

```python
import duckdb

con = duckdb.connect()
csv_path = "../data/raw/population_africa_sample.csv"
con.execute(f"CREATE OR REPLACE VIEW pop AS SELECT * FROM read_csv_auto('{csv_path}')")
con.sql("SELECT * FROM pop LIMIT 5").df()
```

```python
con.sql("""
  SELECT region,
         COUNT(*) AS n_countries,
         SUM(population_millions) AS total_pop_m,
         ROUND(AVG(gdp_per_capita_usd), 0) AS avg_gdp_per_capita
  FROM pop
  GROUP BY region
  ORDER BY total_pop_m DESC
""").df()
```

```python
con.sql("""
  SELECT country, population_millions, gdp_per_capita_usd
  FROM pop
  WHERE gdp_per_capita_usd > 2000
  ORDER BY gdp_per_capita_usd DESC
""").df()
```
